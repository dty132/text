```python
#更新合并
import requests
import csv
import re

# 自定义请求头
headers = {
    'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36",
    'Accept-Language': 'en-US,en;q=0.9'
}

# 打开 CSV 文件并初始化写入器
with open('wood.csv', 'a', encoding='utf_8_sig', newline='') as fp:
    fieldnames = ['name', 'price', 'unit', 'supermaket', 'time']
    writer = csv.DictWriter(fp, fieldnames=fieldnames)
    
    # 如果是新文件，写入表头
    if fp.tell() == 0:
        writer.writeheader()

    # 分页抓取数据
    for x in range(1, 5):
        url = f'https://www.yuzhuprice.com/PorudctPrice?page.curPage={x}'  # 修正了 URL 的格式
        
        try:
            # 设置 verify=False 来忽略 SSL 证书验证
            response = requests.get(url, headers=headers, timeout=10, verify=False)
            response.raise_for_status()  # 检查是否有 HTTP 错误
            html = response.text
            
            # 这里可以使用 lxml 或 BeautifulSoup 解析 HTML 内容
            # 假设 all_tr 是解析后的所有表格行（tr）元素
            
            # 解析 HTML 内容，示例代码
            from lxml import etree
            tree = etree.HTML(html)
            all_tr = tree.xpath('//table//tr')  # 替换为实际的 XPath 表达式
            
            for tr in all_tr:
                tr_data = {
                    'name': re.sub(r'\s+', ' ', ''.join(tr.xpath('./td[1]/text()')).strip()),
                    'price': ''.join(tr.xpath('./td[2]/text()')).strip(),
                    'unit': ''.join(tr.xpath('./td[3]/text()')).strip(),
                    'supermaket': ''.join(tr.xpath('./td[4]/text()')).strip(),
                    'time': ''.join(tr.xpath('./td[5]/text()')).strip()
                }
                print(tr_data)  # 打印数据用于调试
                writer.writerow(tr_data)  # 写入 CSV 文件

        except requests.exceptions.RequestException as e:
            print(f"请求失败: {str(e)}")

```
